```python
import matplotlib.pyplot as plt
import numpy as np

multiplos_de_pi = np.linspace(0, 7*np.pi, 100)
plt.title("Senos y Cosenos")
senos = np.sin(multiplos_de_pi)
cosenos = np.cos(multiplos_de_pi)

plt.subplot(2, 3, 1)
plt.plot(senos, cosenos, 'r')

x = np.array([7, 1, 9, 3])
y = np.array([15, 10, 8, 5])

plt.subplot(2, 3, 2)
plt.plot(x, y, 'g')

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x, y, 'b')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x, y, 'y')

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x, y, 'm')

x = np.array([0, 4, 2, 3])
y = np.array([12, 9, 3, 20])

plt.subplot(2, 3, 6)
plt.plot(x, y, 'c')

plt.show()
```

```python
f1 = lambda x:np.cos(x)
f2 = lambda x:np.sin(x)

x = np.linspace(0,6*np.pi,100)
plt.subplot(1,2,1)
plt.plot(x,f1(x),'1:r')
plt.title("Coseno")

plt.subplot(1,2,2)
plt.plot(x,f2(x),'+:m')
plt.title("seno")

plt.subtitle("Funciones Trigonométricas")

plt.show()
```
