1.- Con base en la definición de número primo, escriba una función que reciba un número entero e imprima 'True' si es primo o de lo contrario imprima 'False'. Tenga en cuenta que si el número es menor o igual a 1, se retorna como resultado "Error en el Número" y debe volver a preguntar hasta que la entrada sea valida.

```python
# Función para verificar si un número es primo
def num_primo(num):
    # Verificar si el número es menor o igual a 1
    if num <= 1:
        print("El número no es mayor o igual a 1. Ingrese otro número.")
        return "Error en el Número"

    #Ver si el número es divisible por maás números que no sean ni el mismo ni uno.
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False

    # Si no hay ningún divisor diferente de 1 y el propio número, entonces es primo
    return True

while True:
    # Solicitar al usuario que ingrese un número entero mayor o igual a 1
    num = int(input("Ingrese un número entero mayor o igual a 1: "))

    # Verificar si el número es primo
    resultado = num_primo(num)

    # Verificar si hay algún error en la entrada
    if resultado == "Error en el Número":
        print(resultado)
    else:
        # Si el número es válido, imprimir si es primo o no y tronar el bucle
        print(resultado)
        break
```

2.- Sin utilizar librerías adicionales cree un arreglo de numéros enteros (entre 0 y 150) y calcule a través de funciones estándar o lambda lo siguiente:

* El promedio
* La varianza
* La desviación estándar
* Visualice los datos en un histograma.
* La suma de los números pares e impares.
* Los múltiplos de 2, 3 y 5

```python
import random
import matplotlib.pyplot as plt

# Crear un arreglo de números enteros aleatorios entre 0 y 150
arr_numeros = [random.randint(0, 150) for _ in range(100)]

promedio = lambda arr: sum(arr) / len(arr)
varianza = lambda arr: sum((x - promedio(arr))**2 for x in arr) / len(arr)
desviacion_estandar = lambda arr: varianza(arr)**0.5

p_res = promedio(arr_numeros)
v_res = varianza(arr_numeros)
d_e_res = desviacion_estandar(arr_numeros)

# Calcular suma de números pares e impares
suma_pares = sum(x for x in arr_numeros if x % 2 == 0)
suma_impares = sum(x for x in arr_numeros if x % 2 != 0)

# Encontrar múltiplos de 2, 3 y 5
multiplos_2 = [x for x in arr_numeros if x % 2 == 0]
multiplos_3 = [x for x in arr_numeros if x % 3 == 0]
multiplos_5 = [x for x in arr_numeros if x % 5 == 0]


# Histograma
font1 = {'family':'serif','color':'darkblue','size':18}
font2 = {'family':'serif','color':'m','size':15}
plt.hist(arr_numeros, bins=30, color='pink', alpha=0.9, edgecolor='m')
plt.title('Histograma de Números Aleatorios', fontdict = font1)
plt.xlabel('Valor', fontdict = font2)
plt.ylabel('Frecuencia', fontdict = font2)
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.3)
plt.show()

print('Promedio: ', p_res)
print('Varianza: ', v_res)
print('Desviación Estándar: ', d_e_res)
print('Suma de Números Pares: ',suma_pares)
print('Suma de Números Impares: ', suma_impares)
print('Múltiplos de 2: ', multiplos_2)
print('Múltiplos de 3: ', multiplos_3)
print('Múltiplos de 5: ', multiplos_5)
```

Apoye a sus compañeros y grafique en python la familia de curvas para la solución de la ecuación diferencial y en un sólo gráfico muestre de diferente color las curvas para los siguientes tres casos:

para valores de C > 0
cuando C = 0
para valores de C < 0
Señale en el gráfico las curvas que corresponden a cada uno de los tres casos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definición de la ecuación diferencial
# solución de la ecuación diferencial y = C*e^t-t-1

def solucion_general(t, C):
    return C * np.exp(t) - t - 1

# Valores de t
t_v = np.linspace(-2, 2, 100)

# C > 0
C_positivo = 1
solucion_positivo = solucion_general(t_v, C_positivo)

# C = 0
C_cero = 0
solucion_cero = solucion_general(t_v, C_cero)

# C < 0
C_negativo = -1
solucion_negativo = solucion_general(t_v, C_negativo)

font1 = {'family':'serif','color':'darkblue','size':18}
font2 = {'family':'serif','color':'red','size':15}
plt.plot(t_v, solucion_positivo, label=f'C > 0 (C={C_positivo})', color='darkblue', linestyle='--')
plt.plot(t_v, solucion_cero, label=f'C = 0 (C={C_cero})', color='m', linestyle=':')
plt.plot(t_v, solucion_negativo, label=f'C < 0 (C={C_negativo})', color='green', linestyle='-.')
plt.xlabel('t', fontdict = font2)
plt.ylabel('y', fontdict = font2)
plt.title('Soluciones de la Ecuación Diferencial', fontdict = font1)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def f(x):
    return (3*x + 7) / (x**2 + 4)

# Rango de valores x
x_vals = np.linspace(-10, 10, 1000)

# Graficar la función
font1 = {'family':'serif','color':'darkblue','size':18}
font2 = {'family':'serif','color':'darkmagenta','size':15}
plt.figure(figsize=(8, 6))
plt.plot(x_vals, f(x_vals), label='f(x) = (3x + 7) / (x^2 + 4)', color='green')

# Graficar las asíntotas horizontales y verticales
plt.axhline(0, color='m', linestyle='--', label='Asíntota Horizontal')
plt.axvline(0, color='purple', linestyle='--', label='Asíntota Vertical en x=0')
plt.ylim(-10, 10)
plt.legend()
plt.xlabel('x', fontdict = font2)
plt.ylabel('f(x)', fontdict = font2)
plt.title('Gráfica de la función con Asíntotas', fontdict = font1)
plt.grid()
plt.show()
```
