```python
SL = df[df['Country']=='Sierra Leone']
```

```python
SL
```

```python
SL = SL.drop('Country',axis=1)
#Para borrar la columna country
```

```python
#Promedio para rellenar el numero de muertes
mean_values = SL['No. of confirmed deaths'].mean()
mean_values
```

```python
SL.fillna(mean_values,inplace=True)
```

```python
SL.isnull().sum()
```

```python
SL['No. of confirmed deaths'].mean()
```

```python
SL['No. of confirmed cases'].mean()
```

```python
Dates = SL.index[0:-1:10]
NCC = SL['No. of confirmed cases'][0:-1:10]
NCD = SL['No. of confirmed deaths'][0:-1:10]
```

```python
fig = plt.figure(figsize=(12,9)) #Para que salga completa la imagen
plt.plot(Dates,NCC,'*-m',ms=5,label ='Infectados')
plt.bar(Dates, NCC, color = "m" ,alpha=0.22)
plt.plot(Dates,NCD, 'o:g', ms=5,label='Decesos')
plt.bar(Dates,NCD, color = "g", alpha = 0.32)
plt.grid()
plt.title('Ã‰bola en Sierra Leona')
plt.ylabel('Casos confirmados')
plt.xlabel('Periodo de tiempo')
plt.xticks(rotation=60, fontsize=6)
plt.legend()
plt.savefig("Ebola_SL.png", dpi=500, format='png')
plt.show()
```

```python
PIvsD = 100*SL['No. of confirmed deaths']/SL['No. of confirmed cases']
SL['PIvsD'] = PIvsD
```

```python
recuperados = ((100-PIvsD)/100)*SL['No. of confirmed cases']
SL['recuperados'] = recuperados
```

```python
SL['PR'] = (100-PIvsD)
```

```python
SL.head()
```

```python
PIvsD_mean = SL['PIvsD'].mean()
PR_mean = SL['PR'].mean()
porcentajes = [PIvsD_mean, PR_mean]
etiquetas = ['PIvsD', 'PR']
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct='%0.1f %%', colors=['m', 'pink'])
font1 = {'family':'serif','color':'blue','size':15}
plt.axis("equal")
plt.title("Porcentaje de Recuperados vs Decesos", fontdict = font1)
plt.show()
```
