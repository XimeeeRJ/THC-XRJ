```python

multiplos_de_pi = np.linspace(0, 7*np.pi, 100)


def grafi(multiplos_de_pi):
  senos = np.sin(multiplos_de_pi)
  cosenos = np.cos(multiplos_de_pi)

  font1 = {'family':'serif','color':'Purple','size':10}
  font2 = {'family':'serif','color':'blue','size':8}
  plt.plot(multiplos_de_pi,senos, '1:m', label='Seno')
  plt.plot(multiplos_de_pi, cosenos, '*-b', label='Coseno')
  plt.xlabel('Mltiplos de Pi (radianes)', fontdict = font2)
  plt.title('Grfico de Seno y Coseno de Mltiplos de Pi', fontdict = font1)
  plt.grid(color = 'black', linestyle = '-.', linewidth = 0.15)
  plt.show()
  return None

print(grafi(multiplos_de_pi))


```
